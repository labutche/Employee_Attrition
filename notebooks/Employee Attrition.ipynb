```python
import pandas as pd

# Load the uploaded IBM HR Analytics Employee Attrition & Performance Dataset
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Display the first few rows to understand the structure of the dataset
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Bar Chart: Distribution of Employee Attrition by Job Satisfaction
sns.countplot(data=df, x='JobSatisfaction', hue='Attrition')
plt.title('Distribution of Employee Attrition by Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Count')
plt.legend(title='Attrition', loc='upper right')
plt.tight_layout()
plt.show()

# Histogram: Years at the Company vs. Attrition
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', multiple='stack', kde=True)
plt.title('Years at the Company vs. Attrition')
plt.xlabel('Years at Company')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Heatmap: Correlation Matrix of Employee Attributes
plt.figure(figsize=(14, 12))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of Employee Attributes')
plt.tight_layout()
plt.show()
```
